```python
# Import required libraries
import sklearn
import pandas as pd
from sklearn import datasets
import numpy as np
```

```python
data_df = pd.read_csv('C:\\Users\\danglc\\Desktop\\machineLearning\\dataSoccer.csv', index_col='Match_ID')
```

```python
data_df = data_df[(data_df['homeOdd'] >= 1.77) & (data_df['homeOdd'] <= 1.95)]
```

```python
data_df
```

```python
data_df['rateAttach'] = data_df['H-Attach']-data_df['A-Attach']
data_df['rateDangerousAttach'] = data_df['H-DangerA']-data_df['A-DangerA']
data_df['rateShot'] = data_df['Home_Shot']-data_df['Away_Shot']
data_df['rateOntarget'] = data_df['H-OnGoal']-data_df['A-OnGoal']
data_df['rateOfftarget'] = data_df['H-OffGoal']-data_df['A-OffGoal']
data_df['rateOddtarget'] = data_df['homeOdd']-data_df['awayOdd']
data_df['rateGoal'] = data_df['H-Score']+data_df['A-Score']
```

```python
# conditions = [
#     (data_df['H-Score'] > data_df['A-Score']),
#     (data_df['H-Score'] == data_df['A-Score']),
#     (data_df['H-Score'] < data_df['A-Score'])]
# values = ['Home', 'Draw', 'Away']
# data_df['1X2Result'] = np.select(conditions, values)

conditions = [
    (data_df['H-Score'] > data_df['A-Score']),
    (data_df['H-Score'] == data_df['A-Score']),
    (data_df['H-Score'] < data_df['A-Score'])]
values = ['Home', 'Away', 'Away']
data_df['1X2Result'] = np.select(conditions, values)
```

```python
data_df = data_df[data_df['H-Attach'].notnull()]
data_df = data_df[data_df['homeOdd'].notnull()]
```

```python
data_df.info()
```

```python
feature = ['League', 'rateAttach', 'rateDangerousAttach','rateShot','rateOfftarget','homeOdd', 'awayOdd', 'drawOdd']
```

```python
# x = data_df.loc[:,feature]
x = data_df[feature].values
# y = data_df['1X2Result'].values
y = data_df['rateGoal'].values
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = ct.fit_transform(x)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x,y, train_size=0.8, test_size=0.2, random_state=0)
```
